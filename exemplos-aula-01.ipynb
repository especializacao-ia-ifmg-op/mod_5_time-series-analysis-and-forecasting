## Ruído Gaussiano

```python
from random import gauss, randrange, random
## Serie gerada por Ruído I.I.D Gaussiano
dados_iid = [gauss(0.0, 1.0) for i in range(500)]
```

```python
serie_iid = pd.Series(dados_iid)
serie_iid.plot()
plt.show()
```

## Random Walk

```python
serie_random_walk = list()
serie_random_walk.append(-1 if random() < 0.5 else 1)
for i in range(1, 1000):
    movement = -1 if random() < 0.5 else 1
    value = serie_random_walk[i-1] + movement
    serie_random_walk.append(value)
```

```python
serie_random_walk = pd.Series(serie_random_walk)
serie_random_walk.plot()
plt.show()
```
